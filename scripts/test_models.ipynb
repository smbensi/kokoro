```python
VOICEPACK = torch.load(f'voices/{VOICE_NAME}.pt', weights_only=True).to(device)
print(f'Loaded voice: {VOICE_NAME}')

```

```python
import numpy as np
from scipy.io.wavfile import write

def save_audio_as_wav(audio_array, sample_rate, filename="audio.wav"):
    
    # ensure data is in the correct format (16-bit integrers)
    if audio_array.dtype != np.int16:
        audio_array = np.int16(audio_array * 32767)
    
    write(filename, sample_rate, audio_array)
    print(f"Saved audio to {filename}")
```

```python
sample_rate = 24000
save_audio_as_wav(audio, sample_rate, "audio.wav")
```
